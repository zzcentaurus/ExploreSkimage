# 1. skimage.data & Display image
skimage.data contains some example images. Let's display them using matplotlib.pylot.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

im = data.camera()

fig, ax = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(8, 4))
plt.subplots_adjust(wspace = 0.1)

ax[0].imshow(im, cmap='gray')
ax[0].set_title('(a) gray colormap')
ax[0].axis('off')
ax[0].set_adjustable('box-forced')

ax[1].imshow(im, cmap='gray', vmin=50, vmax=200)
ax[1].set_title('(b) gray in [50, 200]')
ax[1].axis('off')
ax[1].set_adjustable('box-forced')

ax[2].imshow(im, cmap='jet')
ax[2].set_title('(c) jet colormap')
ax[2].axis('off')
ax[2].set_adjustable('box-forced')
```
