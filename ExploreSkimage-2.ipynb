# 2. Dimensions & Read and display 3D image
## 2.1 Dimentions

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
im = data.camera()
print 'im.dtype = %s' % im.dtype
print 'im.shape = %s' % (im.shape,)
print 'im.ndim = %s' % im.ndim
print 'im.size = %s' % im.size
fig, ax = plt.subplots()
ax.imshow(im, cmap='gray')
ax.axis('off')
ax.set_adjustable('box-forced')
ax.set_title('skimage.data.camera')
```

```python
im = data.chelsea()
print 'im.dtype = %s' % im.dtype
print 'im.shape = %s' % (im.shape,)
print 'im.ndim = %s' % im.ndim
print 'im.size = %s' % im.size
fig, ax = plt.subplots()
ax.imshow(im)
ax.axis('off')
ax.set_adjustable('box-forced')
ax.set_title('skimage.data.chelsea')
txt = 'The image shape is given by (rows, columns, 3).\nHere the Z-slice is 3 representing RGB color channels.'
plt.text(0, 340, txt)
```

## 2.2 TIFF images in 2D & 3D
I'm mostly working on tiff images both in 2D and 3D, and I find the module skimage.external.tifffile quite useful. Let's see how it works.

```python
from skimage.external import tifffile
im = tifffile.imread('/Users/zhouzhou/Dropbox/App/spyder/ExploreSkimage/NOR_RC053_#1.tif')
print 'im.dtype = %s' % im.dtype
print 'im.shape = %s' % (im.shape,)
print 'im.ndim = %s' % im.ndim
print 'im.size = %s' % im.size

# default imshow show image in rgb colormap
tifffile.imshow(im)
tifffile.imshow(im, cmap='gray')
tifffile.imshow(im, vmin=30000, vmax=60000, cmap='gray')
```

```python
im = tifffile.imread('/Users/zhouzhou/Dropbox/App/spyder/ExploreSkimage/DSM_reconstructed_crop-2.tif')
print 'im.dtype = %s' % im.dtype
print 'im.shape = %s' % (im.shape,)
print 'im.ndim = %s' % im.ndim
print 'im.size = %s' % im.size
print 'NOTE the shape!!!'

tifffile.imshow(im, cmap='gray')
txt = 'A slider for all slices.'
plt.text(0, 1.5, txt)
```

We can also display a specific slice using matplotlib.

```python
fig, ax = plt.subplots()
ax.imshow(im[200,:,:], cmap='gray')
ax.axis('off')
ax.set_adjustable('box-forced')
ax.set_title('slice 201')
```

```python
# Look through slices from different directions easily
tifffile.imshow(im.transpose(1,2,0), cmap='gray')
tifffile.imshow(im.transpose(2,1,0), cmap='gray')
```

## 2.3 Z-projection

```python
im_Zmean = np.mean(im, axis=0) # now the dtype is float64
im_Zmax = np.amax(im, axis=0)
im_Zmin = np.amin(im, axis=0)

fig, ax = plt.subplots(1,3)
plt.rcParams['figure.figsize'] = (20, 10)
ax[0].imshow(im_Zmean, vmin=0, vmax=150, cmap='gray')
ax[0].axis('off')
ax[0].set_adjustable('box-forced')
ax[0].set_title('Z mean projection')
ax[1].imshow(im_Zmax, vmin=0, vmax=150, cmap='gray')
ax[1].axis('off')
ax[1].set_adjustable('box-forced')
ax[1].set_title('Z max projection')
ax[2].imshow(im_Zmin, vmin=0, vmax=150, cmap='gray')
ax[2].axis('off')
ax[2].set_adjustable('box-forced')
ax[2].set_title('Z min projection')
```
